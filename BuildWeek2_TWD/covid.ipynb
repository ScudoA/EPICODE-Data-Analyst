### Lettura del csv!

```python

#morti per regione
from warnings import filterwarnings
filterwarnings('ignore')
import pandas as pd
from matplotlib import pyplot as plt

file_name = 'Untitled Folder 2/covid19_italy_region _python.csv'#jupyter vuole il path a differenza di vs code
df = pd.read_csv(file_name)

df
```

```python
df= df.drop(columns=['TestsPerformed'], axis = 1)
```

### Data set Covid Regioni 2020

```python
df
```

### Media morti per Regioni/ Creazione DataFrame

```python
#calcolo la media di morti per regioni

df_morti=df[['Deaths','RegionName']]
```

### Tabella Regioni e Morti

```python
df_morti
```

### Raggruppamento della tabella per Regioni e Media/ Calcolo

```python
#raggruppo per regioni e poi faccio la media delle morti

df_morti_media = df_morti.groupby(['RegionName'])['Deaths'].mean()
df_morti_media1=df_morti_media.sort_values()
```

### Media Morti per Regione/ Risultato

```python
df_morti_media1
```

### Definizione delle variabili X e Y per il grafico

```python
#tracciamo un grafichetto forse meglio chart
x_values = df_morti_media1.index
y_values = df_morti_media1.values
```

# Grafico Morti per Regione/ ordine: ASC

```python
plt.barh(x_values, y_values, color='red')
plt.figure(figsize=(10, 6)) 
plt.tight_layout()
```

### Analisi temporale del totale dei positivi/Somma

```python
#adesso facciamo un'analisi temporale del totale dei positivi
import pandas as pd

df['Date']= pd.to_datetime(df['Date'])
df['Month']= df['Date'].dt.month
df_per_month= df.groupby('Month')['TotalPositiveCases'].sum()
```

```python
df_per_month
```

### Creazione del grafico / casi positivi totali

```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

plt.figure(figsize=(10, 6))  # Impostazione delle dimensioni del grafico
sns.set_theme(style="darkgrid")  # Impostazione dello stile di seaborn

# Plot
sns.lineplot(data=df_per_month, marker='o')  

# Aggiunta di etichette e titoli
plt.title('Casi positivi totali per mese')
plt.xlabel('Mese')
plt.ylabel('Casi positivi totali')

# Mostrare il grafico
plt.show()
```

```python
#morti per regione

import pandas as pd
from matplotlib import pyplot as plt

file_name = 'Untitled Folder 2/covid19_italy_province _python.csv'#jupyter vuole il path a differenza di vs code
df2 = pd.read_csv(file_name)

df2
```

### Pulizia del dataset

```python
import pandas as pd

df2= df2.dropna()

df2
```

### Creazione mappa geografica con il data set delle province
##### Possiamo notare come i dati non siano disordinati ma corrispondenti tra i data set. Poichè la rappresentazione grafica della cartina coincide con quella del dataset precedente

```python
import folium 
import pandas as pd 
from folium.plugins import HeatMap
mappa = folium.Map(location=[df2['Latitude'].mean(), df2['Longitude'].mean()], zoom_start=5)
heat_map = HeatMap(data=df2[['Latitude', 'Longitude', 'TotalPositiveCases']], radius=15)
mappa.add_child(heat_map)
```

### Creazione dataframe Regione Sicilia

```python
#adesso in particolare grazie al dataset COVID19 SICILIA vado ad analizzare l'analisi della regione.
import pandas as pd
df_sicilia = df2[df2['RegionName']== 'Sicilia']
df_sicilia
```

### datetime e raggruppo per mese del totale dei positivi

```python

df_sicilia['Date']= pd.to_datetime(df_sicilia['Date'])
df_sicilia['Month']= df_sicilia['Date'].dt.month

df_sicilia_province_tot = df_sicilia.groupby('Month')['TotalPositiveCases'].sum()
df_sicilia_province_tot
```

### Creazione grafico/ Andamento annuale positivi Regione Sicilia per  provincia 

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Creiamo una figura e un'area degli assi
plt.figure(figsize=(12, 8))

# Utilizziamo seaborn per un aspetto migliore del grafico
sns.set_style("whitegrid")

# Loop attraverso le province della Sicilia e tracciamo l'andamento dei casi positivi per ciascuna
for provincia in df_sicilia['ProvinceName'].unique():
    # Selezioniamo solo i dati relativi alla provincia corrente
    df_province_corrente = df_sicilia[df_sicilia['ProvinceName'] == provincia]
    
    # Raggruppiamo per mese e sommiamo i casi positivi
    df_province_corrente_mensile = df_province_corrente.groupby('Month')['TotalPositiveCases'].sum()
    
    # Plot per la  serie temporale per la provincia corrente
    plt.plot(df_province_corrente_mensile.index, df_province_corrente_mensile.values, label=provincia)

# Aggiungiamo etichette e titoli
plt.title('Andamento dei casi positivi per città nella regione Sicilia')
plt.xlabel('Mese')
plt.ylabel('Numero totale di casi positivi')

# Aggiungiamo la legenda
plt.legend()

# Mostrare il grafico
plt.grid(True)
plt.show()

```

### Totale positivi delle province della Sicilia

```python
df_sicilia_province_tot = df_sicilia.groupby('ProvinceName')['TotalPositiveCases'].sum()
df_sicilia_province_tot
```

### Creazione grafico a barre/ numero totale di casi positivi per provincia

```python
import matplotlib.pyplot as plt

# Ordina i dati per il numero totale di casi positivi
df_sicilia_province_tot_sorted = df_sicilia_province_tot.sort_values()

# Creiamo il grafico a barre
plt.figure(figsize=(10, 6))
plt.bar(df_sicilia_province_tot_sorted.index, df_sicilia_province_tot_sorted.values, color='skyblue')

# Aggiungiamo etichette e titoli
plt.title('Numero totale di casi positivi per provincia nella regione Sicilia')
plt.xlabel('Provincia')
plt.ylabel('Numero totale di casi positivi')

# Ruotiamo le etichette sull'asse x per una migliore leggibilità finalmente sono riuscito a farle vedere
plt.xticks(rotation=45, ha='right')

# Mostrare il grafico
plt.tight_layout()
plt.show()
```
